```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('/content/drive/MyDrive/LG_AIMERS/train (1).csv')
test = pd.read_csv('/content/drive/MyDrive/LG_AIMERS/test (1).csv')
submission = pd.read_csv('/content/drive/MyDrive/LG_AIMERS/submission.csv')
```

```python
drop_cols = []
for i in range(len(train.columns)):
    if len(train.iloc[:, i].unique()) == 1:
        drop_cols.append(train.columns[i])
print(len(drop_cols))

train = train.drop(drop_cols, axis=1)
```

```python
drop_cols = []
for column in train.columns:
  if (train[column].notnull().sum() // 5) < train[
      column].isnull().sum():
      drop_cols.append(column)
train[drop_cols] = train[drop_cols].fillna(0)
train = train
train
```

```python
train = train.T.drop_duplicates().T
train
```

```python
#유니크 하나인 열 드랍
train = train[[col for col in train.columns if train[col].nunique() > 1]]
train
```

```python
# 열의 50% 이상이 결측치인 열 드롭
threshold = len(train) * 0.5
train = train.dropna(thresh=threshold, axis=1)
train
```

```python
test = test[train.columns]
test
```

```python
# 'CURE END POSITION'이라는 문자열이 포함된 열들의 목록 가져오기
cure_end_position_cols_train = [col for col in train.columns if 'CURE END POSITION' in col]
cure_end_position_cols_test = [col for col in test.columns if 'CURE END POSITION' in col]

# 새로운 'CURE END POSITION' 열의 위치 찾기 (train)
cols_train = list(train.columns)
cure_end_position_idx_train = next((i for i, col in enumerate(cols_train) if 'CURE END POSITION' in col), None)

# 새로운 'CURE END POSITION' 열의 위치 찾기 (test)
cols_test = list(test.columns)
cure_end_position_idx_test = next((i for i, col in enumerate(cols_test) if 'CURE END POSITION' in col), None)

# 해당 위치에 새로운 'CURE END POSITION' 열 추가 (train)
if cure_end_position_idx_train is not None:
    train.insert(cure_end_position_idx_train, 'CURE END POSITION', train[cure_end_position_cols_train].sum(axis=1))
else:
    train['CURE END POSITION'] = train[cure_end_position_cols_train].sum(axis=1)

# 해당 위치에 새로운 'CURE END POSITION' 열 추가 (test)
if cure_end_position_idx_test is not None:
    test.insert(cure_end_position_idx_test, 'CURE END POSITION', test[cure_end_position_cols_test].sum(axis=1))
else:
    test['CURE END POSITION'] = test[cure_end_position_cols_test].sum(axis=1)

# 기존 'CURE END POSITION' 관련 열들 삭제 (train)
train = train.drop(cure_end_position_cols_train, axis=1)

# 기존 'CURE END POSITION' 관련 열들 삭제 (test)
test = test.drop(cure_end_position_cols_test, axis=1)
```

```python
def get_crosstab(df, feature):
    crosstab = pd.crosstab(df[feature], df['target'], normalize='index')*100
    crosstab = crosstab.reset_index()
    return crosstab
```

```python
# 결과를 저장할 빈 리스트를 만듭니다.
selected_columns = []

# train 데이터 프레임의 모든 열에 대해 반복합니다.
for column in train.columns:
    if column != 'target':
        crosstab = get_crosstab(train, column)

        # AbNormal 비율이 20% 이상인 행이 있는지 확인합니다.
        if (crosstab['AbNormal'] >= 20).any():
            selected_columns.append(column)

# 선택된 열들을 출력합니다.
print("AbNormal 비율이 20% 이상인 값이 있는 열들:", selected_columns)
```

```python
# AbNormal 비율이 20% 이상인 값을 찾기 위한 교차표 함수
def get_high_abnormal_columns(df, feature, threshold=20):
    crosstab = pd.crosstab(df[feature], df['target'], normalize='index') * 100
    high_abnormal_values = crosstab[crosstab['AbNormal'] >= threshold].index.tolist()
    return high_abnormal_values

# 조건을 충족하는 값들을 찾음
high_abnormal_values = {}
columns_to_check = selected_columns

for column in columns_to_check:
    high_abnormal_values[column] = get_high_abnormal_columns(train, column)

# 조건을 충족하는 열의 값들이 5개 이상인 경우를 찾아서 High_AbNormal_Ratio 열에 1 또는 0을 할당
def check_high_abnormal_ratio(row):
    count = 0
    for column in columns_to_check:
        if row[column] in high_abnormal_values[column]:
            count += 1
    return 1 if count >= 5 else 0

# apply 함수를 사용하여 조건에 맞는 값을 계산하고 새로운 열에 할당
train['High_AbNormal_Ratio'] = train.apply(check_high_abnormal_ratio, axis=1)
test['High_AbNormal_Ratio'] = test.apply(check_high_abnormal_ratio, axis=1)
```
